I need to make sure the "Group" columns corresponds to the number of possible answers.

```python
import pandas as pd
from pathlib import Path


def update_group_from_options(csv_file, output_file=None):
    """
    Updates the 'Group' column using the succession of the last letter
    of each non-empty value in columns starting with 'Option '.
    """

    csv_file = Path(csv_file)

    df = pd.read_csv(csv_file)

    # Get columns that start with "Option "
    option_cols = [col for col in df.columns if col.startswith("Option ")]

    if not option_cols:
        raise ValueError("No columns starting with 'Option ' were found.")

    def build_group(row):
        letters = []

        for col in option_cols:
            value = row[col]
            # print(col)
            # Skip NaN or empty values
            # step 1
            if pd.isna(value):
                continue

            # step 2
            value = str(value).strip()
            if value == "":
                continue

            letters.append(col[-1])
        # print(letters)
        return "".join(letters)

    df["Group"] = df.apply(build_group, axis=1)

    # Save to same file unless output_file is provided
    if output_file is None:
        output_file = csv_file

    df.to_csv(output_file, index=False)
    return df
```

```python
# first I clean a sample from the test split, and make sure experiments are running correcltly.
path_original_test = "data/MedAraBench/medarabench-test_split.csv"
path_cleaned_test = "data/cleaned/medarabench_test.csv"

update_group_from_options(path_original_test, path_cleaned_test)
```
